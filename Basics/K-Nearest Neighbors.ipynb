```python
#Importing the required packages
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
#importing dataset and generating a plot
dataset=pd.read_csv('../Data/KNN/KNNAlgorithmDataset.csv')
dataset.fillna(0, inplace=True)
ds=dataset.to_numpy()
x=ds[:,:-1]
print(x)
y=ds[:,-1]
print(y)
plt.figure(figsize = (10,10))
plt.scatter(x[:,0], x[:,1], c=y, marker= '*',s=100)
plt.show()
```

```python
#Splitting the training data and Testing Data
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
#generating model and Fitting the data
knn5 = KNeighborsClassifier(n_neighbors = 2)
knn5.fit(X_train, y_train)
```

```python
# Prediction of Data
y_pred_5 = knn5.predict(X_test)
```

```python
#Finding Accuracy of the model
from sklearn.metrics import accuracy_score
print("Accuracy with k=5", accuracy_score(y_test, y_pred_5)*100)
```

```python
#Plotting the points
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred_5, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=5", fontsize=20)
plt.show()
```
